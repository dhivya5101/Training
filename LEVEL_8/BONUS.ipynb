```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) 
pareto_data = np.random.pareto(a=2, size=1000)

plt.hist(pareto_data, bins=50, density=True, alpha=0.6, color='g', edgecolor='black', log=True)
plt.xlabel('Value')
plt.ylabel('Probability Density (log scale)')
plt.title('Pareto Distribution (shape=2)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

pareto_data_sorted = np.sort(pareto_data)  
top_20_percent = pareto_data_sorted[-200:] 
total_sum = np.sum(pareto_data)
top_20_sum = np.sum(top_20_percent)

pareto_rule_verified = (top_20_sum / total_sum) >= 0.8
verification_result = "Top 20% of values account for â‰ˆ 80% of total sum, confirming the 80/20 rule." if pareto_rule_verified else "80/20 rule not exactly met, but trend is visible."

top_20_sum / total_sum, verification_result
```
