```python
from dataclasses import dataclass
import pandas as pd
import matplotlib.pyplot as plt

@dataclass
class DataStats:
    mean: float
    max_value: float
    min_value: float
    
    def plot_histogram(self, df, column_name):
        plt.hist(df[column_name], bins=30, color='blue', alpha=0.7)
        plt.xlabel(column_name)
        plt.ylabel("Frequency")
        plt.title(f"Histogram of {column_name}")
        plt.grid(True)
        plt.show()
```

```python
df = pd.read_csv("house_prices.csv")
df
```

```python
column_name = "Price"
mean_val = df[column_name].mean()
max_val = df[column_name].max()
min_val = df[column_name].min()

data_stats = DataStats(mean=mean_val, max_value=max_val, min_value=min_val)
```

```python
while True:
    user_input = input("Enter 'stats' to view statistics or 'plot' to see histogram (or 'exit' to quit): ").strip().lower()
    
    if user_input == "stats":
        print(f"\nStatistics for {column_name}:")
        print(f"Mean: {data_stats.mean}")
        print(f"Max: {data_stats.max_value}")
        print(f"Min: {data_stats.min_value}\n")
    
    elif user_input == "plot":
        data_stats.plot_histogram(df, column_name)
    
    elif user_input == "exit":
        print("Exiting...")
        break
    else:
        print("Invalid input. Please enter 'stats', 'plot', or 'exit'.")
```
