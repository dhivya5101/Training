```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("house_prices.csv")  
df
```

```python
df = df.dropna(subset=['BuildingArea'])
```

```python
summary_stats = df[['Price', 'BuildingArea']].describe()
print("Summary Statistics:\n", summary_stats)
```

```python
df['price_per_sqft'] = df['Price'] / df['BuildingArea']
print("\nUpdated DataFrame:\n", df[['Price', 'BuildingArea', 'price_per_sqft']].head())
```

```python
plt.scatter(df['BuildingArea'], df['Price'], alpha=0.5, color='blue')
plt.xlabel("Building Area (sq meters)")
plt.ylabel("Price")
plt.title("House Price vs. Building Area")
plt.grid(True)
plt.show()
```
