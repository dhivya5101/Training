```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
df
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
df
```

```python
if 'diagnosis' not in df.columns:
    raise ValueError("Error: 'diagnosis' column not found in dataset.")

y = df['diagnosis'].astype(str) 
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

X = df.drop(columns=['diagnosis'], errors='ignore')
X = X.apply(pd.to_numeric, errors='coerce') 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
print("Data preprocessing completed successfully!")
```

```python
base_model = DecisionTreeClassifier(max_depth=1, random_state=42)
base_model.fit(X_train, y_train)
y_pred_base = base_model.predict(X_test)
```

```python
adaboost_model = AdaBoostClassifier(base_model, n_estimators=50, random_state=42)
adaboost_model.fit(X_train, y_train)
y_pred_adaboost = adaboost_model.predict(X_test)

accuracy_base = accuracy_score(y_test, y_pred_base)
accuracy_adaboost = accuracy_score(y_test, y_pred_adaboost)
roc_auc_base = roc_auc_score(y_test, y_pred_base)
roc_auc_adaboost = roc_auc_score(y_test, y_pred_adaboost)
```

```python
comparison_df = pd.DataFrame({
    "Model": ["Decision Tree (Base)", "AdaBoost"],
    "Accuracy": [accuracy_base, accuracy_adaboost],
    "ROC-AUC": [roc_auc_base, roc_auc_adaboost]
})

print("\n Model Performance Comparison:")
print(comparison_df)
```

```python
plt.figure(figsize=(6, 4))
plt.bar(["Decision Tree", "AdaBoost"], [accuracy_base, accuracy_adaboost], color=['red', 'blue'])
plt.ylabel("Accuracy Score")
plt.title("Base vs. AdaBoost Performance")
plt.ylim(0.8, 1.0)
plt.show()
```
